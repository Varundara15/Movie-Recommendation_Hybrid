```python
from sklearn.metrics.pairwise import cosine_similarity
# movie_indices = pd.Series(final_movies.index, index=final_movies['title']).drop_duplicates()
# def get_similar_movies(movie_title, n=10):
#     idx = movie_indices[movie_title]

#     # Compute similarity only for the given movie
#     sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()

#     # Get top N similar movies
#     similar_movies = sim_scores.argsort()[-(n+1):-1][::-1]
#     return final_movies.iloc[similar_movies][['title']]
```

```python
movie_indices = pd.Series(movies.index, index=movies["title"]).drop_duplicates()
def get_similar_movies(movie_title, n=10):
    if movie_title not in movie_indices:
        return pd.DataFrame(columns=["movieId", "title", "tmdbId"])

    idx = movie_indices[movie_title]
    sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()

    # Find top N similar movies
    similar_movies = np.argsort(sim_scores)[::-1][1:n+1]

    return movies.iloc[similar_movies][["movieId", "title", "tmdbId"]]
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
reader = Reader(rating_scale=(0.5, 5.0))
```

```python
# Load the dataset into Surprise
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

# Split the data into train and test sets
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
ratings_sample = ratings.sample(frac=0.3, random_state=42)  # Reduce dataset size

# Define reader and dataset
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(ratings_sample[['userId', 'movieId', 'rating']], reader)

# Build full trainset
trainset = data.build_full_trainset()

# Manually set best hyperparameters
best_svd = SVD(n_factors=50, n_epochs=20, lr_all=0.005, reg_all=0.1)

# Train model
best_svd.fit(trainset)

# Evaluate
testset = trainset.build_testset()
predictions = best_svd.test(testset)
rmse(predictions)
```

```python
# Step 2: Predict ratings for similar movies using SVD
def get_svd_predictions(user_id, movie_ids):
    predictions = best_svd.test([(user_id, movie_id, 0) for movie_id in movie_ids])
    return pd.DataFrame([(pred.iid, pred.est) for pred in predictions], columns=["movieId", "predicted_rating"])
```

```python
# Get all unique movie IDs
all_movie_ids = ratings["movieId"].unique()

# Select a specific user
user_id = 1

# Get movies the user has already rated
watched_movie_ids = ratings[ratings["userId"] == user_id]["movieId"].tolist()

# Filter out watched movies
unwatched_movies = np.setdiff1d(all_movie_ids, watched_movie_ids)  # Faster than list comprehension

# Predict ratings for all unwatched movies (Batch Processing)
predictions = best_svd.test([(user_id, movie_id, 0) for movie_id in unwatched_movies])

# Sort movies by predicted rating
top_movies = sorted(predictions, key=lambda x: x.est, reverse=True)[:10]

# Display recommendations
recommended_movies = pd.DataFrame(
    [(pred.iid, movies[movies["movieId"] == pred.iid]["title"].values[0], pred.est) for pred in top_movies],
    columns=["movieId", "title", "predicted_rating"]
)

```

```python
def get_hybrid_recommendations(user_id, movie_title, alpha=0.5, num_recs=10):
    # Get similar movies from content-based filtering
    content_recs = get_similar_movies(movie_title, num_recs)

    if content_recs.empty:
        return pd.DataFrame(columns=["movieId", "title", "tmdbId", "final_score"])

    # Get SVD-based predictions for those movies
    svd_recs = get_svd_predictions(user_id, content_recs["movieId"].tolist())

    # Merge both based on movieId
    hybrid_df = content_recs.merge(svd_recs, on="movieId", how="left")

    # Fill missing predicted ratings with average rating
    avg_rating = svd_recs["predicted_rating"].mean() if not svd_recs.empty else 3.0
    hybrid_df["predicted_rating"] = hybrid_df["predicted_rating"].fillna(avg_rating)

    # Assign content similarity score
    hybrid_df["content_score"] = np.linspace(1, 0, len(hybrid_df))

    # Final Score Calculation
    hybrid_df["final_score"] = alpha * hybrid_df["predicted_rating"] + (1 - alpha) * hybrid_df["content_score"]

    # Sort by final score
    hybrid_df = hybrid_df.sort_values(by="final_score", ascending=False)

    return hybrid_df[["movieId", "title", "tmdbId", "final_score"]].head(10)

```

```python
user_id = 1
movie_title = "Toy Story (1995)"

recommended_movies = get_hybrid_recommendations(user_id, movie_title)
print(recommended_movies)
```
